```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.zeros((100,50))
b = np.ones((100,50))
```

```python
img = np.uint8(255*np.concatenate((a,b), axis=1))
```

```python
cv2.imshow('imgen 0 y 1',img)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(img)
plt.title("Original")
```

```python
Gx = cv2.Sobel(img, cv2.CV_64F, 1, 0, 5)
cv2.imshow('Gx',Gx)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(Gx)
plt.title("Gx")
```

```python
Gy = cv2.Sobel(img, cv2.CV_64F, 0, 1, 5)
cv2.imshow('Gy',Gy)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(Gy)
plt.title("Gy")
```

```python
Mag, Ang = cv2.cartToPolar(Gx, Gy)
cv2.imshow('Mag',Mag)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(Mag)
plt.title("Magnitud Gx^2+Gy^2")
```

```python
Gx = cv2.convertScaleAbs(Gx)
Gy = cv2.convertScaleAbs(Gy) 
Mag = cv2.convertScaleAbs(Mag)
Ang = (180/np.pi)*Ang
```

```python
img = cv2.imread('./img/bananos.jpg',0)
Gx = cv2.Sobel(img, cv2.CV_64F, 1, 0, 5)
Gy = cv2.Sobel(img, cv2.CV_64F, 0, 1, 5)
Mag, Ang = cv2.cartToPolar(Gx, Gy)

Gx = cv2.convertScaleAbs(Gx)
Gy = cv2.convertScaleAbs(Gy) 
Mag = cv2.convertScaleAbs(Mag)
Ang = (180/np.pi)*Ang

#Filtro Gausiano 
imgFilt = cv2.GaussianBlur(img, (5,5),0)
#Laplaciano 
lap = cv2.convertScaleAbs(cv2.Laplacian(imgFilt, cv2.CV_64F, 5))
```

```python
cv2.imshow('Gx',Gx)
cv2.imshow('Gy',Gy)
cv2.imshow('Mag',Mag)
cv2.imshow('Lap',lap)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(Gx)
plt.title(" Gx")
plt.show()
plt.imshow(Gy)
plt.title("Gy")
plt.show()
plt.imshow(Mag)
plt.title("Magnitud Gx^2+Gy^2")
plt.show()
plt.imshow(lap)
plt.title("Laplaciano")
plt.show()
```

```python
# Deteccion de bordes con Canny
canny=cv2.Canny(img,25,150)
```

```python
cv2.imshow('Gx',Gx)
cv2.imshow('Gy',Gy)
cv2.imshow('Mag',Mag)
cv2.imshow('Lap',lap)
cv2.imshow('Canny',canny)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(Gx)
plt.title(" Gx")
plt.show()
plt.imshow(Gy)
plt.title("Gy")
plt.show()
plt.imshow(Mag)
plt.title("Magnitud Gx^2+Gy^2")
plt.show()
plt.imshow(lap)
plt.title("Laplaciano")
plt.show()
plt.imshow(canny)
plt.title("Canny")
plt.show()
```

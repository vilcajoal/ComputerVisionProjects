```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./img/bananos.jpg')
```

```python
img.shape
```

```python
cv2.imshow('Original',img)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(img)
```

```python
kernel_3x3 = np.ones((3,3), np.float32)/(3*3)
output = cv2.filter2D(img, -1, kernel_3x3)
cv2.imshow('PRomedio 3x3',output)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(output)
```

```python
kernel_5x5 = np.ones((5,5), np.float32)/(5*5)
output = cv2.filter2D(img, -1, kernel_5x5)
cv2.imshow('PRomedio 5x5',output)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(output)
```

```python
kernel_31x31 = np.ones((31,31), np.float32)/(31*31)
output = cv2.filter2D(img, -1, kernel_31x31)
cv2.imshow('PRomedio 31x31',output)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(output)
```

```python
output = cv2.GaussianBlur(img, (3,3), 0)
cv2.imshow('Gaussian Dev = 3x3',output)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(output)
```

```python
output = cv2.GaussianBlur(img, (11,11), 0)
cv2.imshow('Gaussian Dev = 11x11',output)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(output)
```

```python
output = cv2.GaussianBlur(img, (21,21), 0)
cv2.imshow('Gaussian Dev = 21x21',output)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(output)
```

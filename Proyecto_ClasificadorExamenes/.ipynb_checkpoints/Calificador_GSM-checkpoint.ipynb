```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import imutils
```

```python
#Leer imagen
img = cv2.imread("IMG_02.png", 0)
img = cv2.resize(img,(1040,1640))
```

```python
# ROTACION 
rows,cols = img.shape
rotated90 = imutils.rotate_bound(img, -90)
```

```python
#Mostrar imagen rotada
cv2.imshow("Exámen",rotated90)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(rotated90)
```

```python
rotated90.shape
```

```python
#Deteccion de bordes con umbral min y max
canny = cv2.Canny(rotated90, 100, 150)
```

```python
#Mostrar imagen con borde canny
cv2.imshow("borde canny",canny)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(canny)
```

```python
#Aumento de bordes mediante kernel
kernel = np.ones((5,5), np.uint8)
bordes = cv2.dilate(canny, kernel)
```

```python
#Mostrar imagen con borde aumentado
cv2.imshow("borde aumentado",bordes)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(bordes)
```

```python
#Deteccion de bordes de poligono de área maxima
_,contour,_ = cv2.findContours(bordes, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
#Pimera mascara llamada objetos
objetos = bordes.copy()
```

```python
#Llenar el objeto grande y de mayor area que es la region de interes 
# -1: detecte
# 255: ponga en blanco
cv2.drawContours(objetos, [max(contour, key=cv2.contourArea)], -1, 255, thickness = -1)
```

```python
#Mostrar imagen con borde aumentado
cv2.imshow("borde llenado",objetos)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(objetos)
```

```python
# etiquetar objetos en la mascara de imagen
output = cv2.connectedComponentsWithStats(objetos, 4, cv2.CV_32S)
```

```python
#Número de objetos de salida
numObj = output[0]
# etiquetas de imagen
labels = output[1]
# datos de la img como area
stats = output[2]
```

```python
mascara = np.uint8(255*(np.argmax(stats[:, 4][1:])+1 == labels))
```

```python
#Mostrar imagen de solo la mascara
cv2.imshow("mascara",mascara)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(mascara)
```

```python
#Deteccion de bordes de poligono de área maxima
_,contours,_ = cv2.findContours(mascara, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
#contornos en pocision cero
cnt = contours[0]
```

```python
#Mejorar contornos de la imagen haciendo la imagen convexa
hull = cv2.convexHull(cnt)
puntosConvex = hull[:, 0, :]
m, n = mascara.shape
ar = np.zeros((m, n))
mascaraConvex = np.uint8(cv2.fillConvexPoly(ar, puntosConvex, 1))
```

```python
#Mostrar imagen de solo la mascara
cv2.imshow("mascara Convexa Mejorada", mascaraConvex)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(mascaraConvex)
```

```python
#Correccion de perrpectiva
#Obtener los vertices del area maxima
vertices = cv2.goodFeaturesToTrack(mascaraConvex, 4, 0.01, 30)
```

```python
#Puntos iniciales 
vertices
```

```python
# Crear homografia
x = vertices[:, 0, 0]
y = vertices[:, 0, 1]
```

```python
# reescalando a matriz de 4x0x2 para la homografia
vertices = vertices[:, 0, :]
```

```python
#Hallar los vertices finales que seran los puntos maximos de la imagen m, n
#Ordenamos los puntos de x y y
Xo = np.sort(x)
Yo = np.sort(y)
```

```python
#Puntos finales
Xn = np.zeros((1,4))
Yn = np.zeros((1,4))
```

```python
#Crear puntos esquinas
Xn = (x == Xo[2])*n + (x ==Xo[3])*n
Yn = (y ==Yo[2])*m +(y==Yo[3])*m
```

```python
#
verticesN = np.zeros((4,2))
verticesN[:, 0] = Xn
verticesN[:, 1] = Yn
```

```python
#pasar a enteros a puntos iniciales y finales
vertices = np.int64(vertices)
verticesN = np.int64(verticesN)
```

```python
# Crear la matriz h de homografia 
h,_ = cv2.findHomography(vertices, verticesN)
```

```python
#Aplicar la matriz de Homografía en la imagen
img_homogra = cv2.warpPerspective(rotated90, h, (n, m))
```

```python
#Mostrar imagen con cambio de perspectiva
cv2.imshow("imagen con cambio de perspectiva", img_homogra)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(img_homogra)
```

```python
def on_EVENT_LBUTTONDOWN(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        xy = "%d,%d" % (x, y)
        cv2.circle(img_homogra, (x, y), 1, (255, 0, 0), thickness = -1)
        cv2.putText(img_homogra, xy, (x, y), cv2.FONT_HERSHEY_PLAIN,
                    1.0, (0,0,0), thickness = 1)
        cv2.imshow("image", img_homogra)
cv2.namedWindow("image")
cv2.setMouseCallback("image", on_EVENT_LBUTTONDOWN)
while(1):
    cv2.imshow("image", img_homogra)
    if cv2.waitKey(0)&0xFF==27:
        break
cv2.destroyAllWindows()
```

```python
opciones=['A','B','C','D','E','x']
respuestas=[]
    
for i in range(0,15):
    pregunta=H_verbal[np.uint64(i*(H_verbal.shape[0]/15)):np.uint64((i+1)*(H_verbal.shape[0]/15)),:]
    sumI=[]
    for j in range(0,5):
        _,col=pregunta.shape
        sumI.append(np.sum(pregunta[:,np.uint64(j*(col/5)):np.uint64((j+1)*(col/5))])) 
    vmin=np.ones((1,5))*np.min(sumI)
    
    if np.sort(sumI-vmin)[0][1]<10000:
        sumI.append(-1)
    elif np.linalg.norm(sumI-vmin)>0.50*col*m: 
        sumI.append(float('inf')) 
    else:sumI.append(-1)
        
    respuestas.append(opciones[np.argmin(sumI)])
```

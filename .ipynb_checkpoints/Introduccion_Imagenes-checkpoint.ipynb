```python
rojo = img[:,:,0].flatten()
verde = img[:,:,1].flatten()
azul = img[:,:,2].flatten()
```

```python
# Histograma de colores en una imagen
plt.hist(rojo, bins = 1000, histtype='stepfilled', color = 'r', range=(0,254))
plt.hist(verde, bins = 1000, histtype='stepfilled', color = 'g', range=(0,254))
plt.hist(azul, bins = 1000, histtype='stepfilled', color = 'b', range=(0,254))
plt.show()
```

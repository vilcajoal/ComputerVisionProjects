```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./img/cuadro.jpg")
```

```python
cv2.imshow('Original',img)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(img)
```

```python
imG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('Escala de Grises',imG)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(imG)
```

```python
m, n, _ = img.shape
```

```python
puntos_iniciales = np.array([[42, 12], [193, 50], [40, 304], [198, 271]])
```

```python
puntos_finales = np.array([[0,0], [n,0], [0,m], [n,m]])
```

```python
h, _ = cv2.findHomography(puntos_iniciales, puntos_finales)
```

```python
img_nueva = cv2.warpPerspective(img, h, (n,m))
```

```python
cv2.imshow('Nueva Perspectiva',img_nueva)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(img)
plt.title("Imagen normal")
plt.show()
plt.imshow(img_nueva)
plt.title("Imagen nueva Perspectiva")
plt.show()
```
